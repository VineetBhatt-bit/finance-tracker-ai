```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
Data={"Value":np.random.randint(1,500,500)
     }
df=pd.DataFrame(Data)
df
```

```python
X = df[['Value']].values
X
```

```python
wcss = []

for k in range(1, 22):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 22), wcss, marker='o')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```

```python
plt.figure()
plt.scatter(df["Value"], [0]*len(df))
plt.title("Before Clustering")
plt.xlabel("Value")
plt.yticks([])
plt.show()
```

```python
kmeans = KMeans(n_clusters=9, random_state=42)
df["Cluster"] = kmeans.fit_predict(X)
centroids = kmeans.cluster_centers_
```

```python
plt.figure()
plt.scatter(df["Value"], [0]*len(df), c=df["Cluster"])
plt.scatter(centroids[:, 0], [0]*len(centroids), marker='h')
plt.title("After Clustering")
plt.xlabel("Value")
plt.show()
```

```python
print("\nClustered Data:")
print(df)
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
df = pd.DataFrame({"Value": np.random.randint(1, 500, 500)})
X = df[['Value']].values
```

```python
sil_scores = []
k_range = range(2, 11)
```

```python
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    sil_scores.append(score)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(k_range, sil_scores, marker='o', color='Blue')
plt.title("Silhouette Method for Optimal K")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.grid(True)
plt.show()
```

```python
optimal_k = k_range[np.argmax(sil_scores)]
print(optimal_k)
```

```python
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df["Cluster"] = kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(10, 4))
jitter = np.random.normal(0, 0.01, size=len(df))
plt.scatter(df["Value"], jitter, c=df["Cluster"], cmap='viridis', alpha=0.5)
plt.scatter(kmeans.cluster_centers_[:, 0], [0] * optimal_k, color='red', marker='X', s=200, label="Centroids")
plt.title(f"Final Clusters (K={optimal_k})")
plt.yticks([])
plt.legend()
plt.show()
```
